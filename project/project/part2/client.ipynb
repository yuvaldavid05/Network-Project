```python
import socket
import threading

SERVER_IP = "127.0.0.1"
PORT = 5000


def recv_loop(sock: socket.socket):
    try:
        f = sock.makefile("r", encoding="utf-8", newline="\n")
        while True:
            line = f.readline()
            if not line:
                print("\n[CLIENT] Server disconnected.")
                break
            print("\n" + line.strip())
    except Exception:
        print("\n[CLIENT] Connection lost.")
    finally:
        try:
            sock.close()
        except Exception:
            pass


def main():
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    sock.connect((SERVER_IP, PORT))

    threading.Thread(target=recv_loop, args=(sock,), daemon=True).start()

    name = input("Enter your name: ").strip()
    sock.sendall((name + "\n").encode("utf-8"))

    print("\nCommands:")
    print("  /chat Bob")
    print("  hello (sends to current peer after /chat)")
    print("  Bob:hello (one-shot)")
    print("  /leave")
    print("  /bye\n")

    while True:
        try:
            msg = input()
        except (EOFError, KeyboardInterrupt):
            msg = "/bye"

        if not msg:
            continue

        try:
            sock.sendall((msg + "\n").encode("utf-8"))
        except Exception:
            print("[CLIENT] Failed to send (connection lost).")
            break

        if msg.lower() in ("/bye", "bye", "exit"):
            break


if __name__ == "__main__":
    main()
```
